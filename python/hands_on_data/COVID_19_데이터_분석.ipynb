## 데이터 결측치 확인

```python
df_patient.isna().sum()
```

나이(age) - 20s -> 23 정확하게 변경

1.   Null(Nan) 데이터를 처리 (Nan 계산 불가) -> 0으로 채우자
2.   나이 계산 시, 0보다 큰 값은 (현재년도 - 태어난 년도) 0보다 작으면 다시 Nan 채우기 



```python
# float 타입에 맞게 넣고 int 형변환
df_patient['birth_year'] = df_patient.birth_year.fillna(0.0).astype(int)
df_patient.tail()
```

```python
df_patient.isna().sum()
```

```python
# 0 보다 큰 값은 그대로 사용하고, 0인 값은 Nan으로 변환 -> 나이 계산시 음수(2021살) 방지
df_patient['birth_year'] = df_patient['birth_year'].map(
    lambda val : val if val >0 else np.nan)
```

```python
df_patient.isna().sum()
```

```python
# Nan 처리한 값으로 처리하고 
df_patient['age'] = 2021 - df_patient['birth_year']
df_patient.tail()
```

```python
patient = df_patient
```

2. 격리해제자(released) 데이터

```python
# 격리해제자 정보 담기
released = patient[patient.state == 'released']
released.info()
```

격리해제자들의 연령대 분석

```python
# 격리해제자들의 연령대 분포
plt.figure(figsize=(10,6), dpi=200)
sns.set_style('darkgrid')
plt.title("Age distribution of released")
sns.kdeplot(data=released['age'], shade=True) # kernel density estimation 확률밀도 추정
```

4. 사망자(deceased) 데이터

```python
# 사망자 정보
dead = patient[patient.state == 'deceased']
dead.info()
```

```python
# 사망자들의 연령대 분포
plt.figure(figsize=(10,6), dpi=200)
sns.set_style('darkgrid')
plt.title("Age distribution of deceased")
sns.kdeplot(data=dead['age'], shade=True)
```

5. 사망자에 따른 성별(gender) 데이터

```python
male_dead = dead[dead.gender == 'male']
female_dead = dead[dead.gender == 'female']
```

```python
male_dead.info()
```

```python
female_dead.info()
```

```python
# 성별에 따른 사망자들의 연령대 분포
plt.figure(figsize=(10,6), dpi=200)
sns.set_style('darkgrid')
plt.title("Age distribution of deceased the gender")
sns.kdeplot(data=male_dead['age'], label='male', shade=True)
sns.kdeplot(data=female_dead['age'], label='female', shade=True)
plt.legend()
```

KDE : Kernel Density Estimater 추가한 distplot()

```python
# 성별에 따른 사망자들의 연령대 분포
plt.figure(figsize=(10,6), dpi=200)
sns.set_style('darkgrid')
plt.title("Age distribution of deceased the gender")
sns.distplot(a=male_dead['age'], label='male', kde=True)
sns.distplot(a=female_dead['age'], label='female', kde=True)
plt.legend()
```

사망자 성별 비교 -> bar chart, pie chart -> 분포

```python
plt.figure(figsize = (10,6), dpi=200)
plt.title('Gender')
dead.gender.value_counts().plot.bar()
```

```python
plt.figure(figsize = (10,6), dpi=200)
plt.title('Gender')
dead.gender.value_counts().plot.pie(autopct = '%.1f%%') # 퍼센테이지도 넣어보자
```
