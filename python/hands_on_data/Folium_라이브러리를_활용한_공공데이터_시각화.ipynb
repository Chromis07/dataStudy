# 세종 어울링 (공공자전거) 시각화

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/세종특별자치시_공영자전거 어울링 대여소_20200915.csv', encoding='cp949')
df.head()
```

```python
df.info()
```

지도에 표시할 대여소 정보 추출

```python
df_route = df.loc[:,['STATION_NM', 'AREA', 'Y_POS', 'X_POS']]
df_route.head()
```

자전고 대여소 Marker 활용하여 지도 표출

```python
# folium 라이브러리
import folium

# 위도, 경도들의 평균값으로 기준점 설정
map = folium.Map(location = [df_route['Y_POS'].mean(), df_route['X_POS'].mean()],
                             zoom_start=11)

for n in df_route.index :
  folium.Marker([df_route.loc[n, 'Y_POS'], df_route.loc[n, 'X_POS']],
                  popup=df_route.loc[n, 'STATION_NM'] + ' ' + df_route.loc[n, 'AREA'],
                icon=folium.Icon(color='green', icon='fa-bicycle', prefix='fa')).add_to(map)

map
```

# 세종시 가로등 설치 데이터 시각화

```python
import pandas as pd
```

```python
lamp_df = pd.read_csv('/content/세종특별자치시_가로등정보_20190919..csv', encoding='cp949')
lamp_df.head()
```

지도에 표시할 가로등 정보 추출

```python
lamp_route = lamp_df.loc[:, ['가로등위치명', '위도', '경도']]
lamp_route.head()
```

가로등 Marker 활용하여 지도 표출

```python
# 가로등 정보 100건만 사용하여 지도 표출
import folium

map = folium.Map(location=[lamp_route['위도'].mean(), lamp_route['경도'].mean()],
                 zoom_start=11)

for n in lamp_route.index[:100] :
  folium.Marker([lamp_route.loc[n, '위도'], lamp_route.loc[n, '경도']],
                 popup=lamp_route.loc[n, '가로등위치명'],
                 icon=folium.Icon(icon='lightbulb-o', prefix='fa')).add_to(map)

map
```

# 세종시 CCTV 설치 데이터 시각화

```python
import pandas as pd
```

```python
cctv_df = pd.read_csv('/content/세종특별자치시_CCTV_20210101.csv', encoding='cp949')
cctv_df.head()
```

```python
# 지도에 표시할 cctv 정보 추출
cctv_route = cctv_df.loc[:, ['카메라화소수', '위도', '경도']]
cctv_route.head()
```

```python
# cctv 정보 100건만 사용하여 지도 표출
import folium

map_cctv = folium.Map(location=[cctv_route['위도'].mean(), cctv_route['경도'].mean()],
                 zoom_start=11)

for n in cctv_route.index[:100]: # 전체 데이터 하려면 대괄호 제거
  folium.Marker([cctv_route.loc[n, '위도'], cctv_route.loc[n, '경도']],
                icon=folium.Icon(icon='video-camera', prefix='fa')).add_to(map_cctv)

map_cctv # 위에 가로등 데이터와 함께 map 표시
```
