```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

1. 에너지 데이터 불러오기

```python
df = pd.read_csv('energy.csv')
df.head()
```

```python
df.info()
```

```python
df.tail()
```

테이블의 첫째 열을 'day'로 변경

```python
df.set_index('day') # 아직 원본이 바뀐건 아님. 변수에 담아줘야함
```

전체 테이블에서 원하는 열만 가져와서 또 다른 파일 만들기

```python
df1 = df[['day', '환기', '콘센트', '냉방', '난방', '조명', '기타']]
df1.head()
```

```python
df1.to_excel('newfile.xlsx') # 전처리한 데이터프레임을 엑셀 파일로 저장
```

2. 값 비교하기

```python
df[df.month == 10] # df[df['month'] == 10]
```

```python
df[df.환기 >=2]
```

3. and, or를 이용해 여러 조건으로 파일 불러오기

```python
df2 = df[(df.month == 10) & (df.weekday == 'Saturday')]
df2.head()
```

자기 생일이 있는 월을 선택하여 냉방이 1보다 크거나 같으면 df4라는 변수에 저장

```python
df3 = df[(df.month == 6) & (df.냉방 >= 1)]
df3.head()
```

```python
df3.to_excel('6월냉방.xlsx')
```
