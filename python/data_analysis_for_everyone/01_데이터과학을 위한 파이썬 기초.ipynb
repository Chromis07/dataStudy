strip() 공백 제거 함수

```python
st = "Hello, world! \t\t\n" # 모든 줄 바꿈과 탭 (\n, \r, \t)
st
```

```python
st.strip() # 모든 공백 제거
```

```python
st.lstrip() # 왼쪽 공백 제거
```

```python
st.rstrip() # 오른쪽 공백 제거
```

split(delim="") 문자열 자르기 함수

```python
"Hello, world!".split() # delim 구분자 -> default 공백
```

```python
"Hello, world!".strip().split(",") # 공백 제거하고 구분자 -> ,(콤마)
```

```python
"www.naver.com".split(".")
```

join(ls) 문자열을 연결시키는 함수

```python
# 문자열 리스트 ls 하나의 문자열로 붙임 -> 문자열 조각을 재조합
", ".join(["alpha", "bravo", "charlie", "delta"])
```

```python
# . 구분자로 잘라 만들어진 리스트를 다시 join으로 재조합
"-".join("1.735.463.3685".split("."))
```

```python
# 1)
ls = "1.735.463.3685".split(".")
# 2)
"-".join(ls)
```

```python
print("This string\n\r has many\t\tspacese.")
```

```python
#공백 전부 없애고 스페이스바 기준으로 합치기
" ".join("This string\n\r has many\t\tspacese.".split())
```

find(needle) 해당 문자열에 부문 문자열 needle 처음 등장하는 인덱스 반환(위치) 없으면 -1 (대소문자 구분)

```python
"www.naver.com".find(".com")
```

```python
"www.naver.Com".find(".com")
```

count(needle) 해당 문자열에서 등장하는 횟수를 반환하는 함수 (대소문자 구분)

```python
"www.naver.com".count(".")
```

```python
"www.naver.com".count("w")
```

```python
"www.naver.com".count("C") # 없을 시엔 0 반환
```

프리디케이트 함수 -> 문자열 s가 특정 클래스에 속하는지 True, False로 반환

```python
"www.naver.com".islower() # 모든 알파벳 문자가 소문자인지 체크
```

```python
"www.naver.com".isupper() # 모든 알파벳 문자가 소문자인지 체크
```

```python
"www.naver.com".isdigit() # 0~9 사이 숫자인지 확인
```

```python
"wwwnavercom".isalpha() # dot 제거해야만  True -> 알파벳(숫자 특수기호 X)인지 체크
```

대소문자 변환 함수 -> 모든 문자 대소문자 변환 (문자열 S의 복사본 반환)

```python
st = "www.naver.com".upper()
st
```

```python
st.lower() # 값을 넣지 않았기에 소문자 변환 안됨
st
```
