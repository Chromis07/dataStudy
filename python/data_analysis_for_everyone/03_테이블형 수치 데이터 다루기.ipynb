데이터 클리닝(data cleaning)을 위해 사용 되는 boolean indexing


```python
# 절대 음수 값이 와서는 안되는 데이터셋
dirty = np.array([9, 4, 1, -0.01, -0.02, -0.001])
whos_dirty = dirty < 0 # boolean indexing - 리스트 내포로 하나씩 꺼내와서 체크할 필요없음
whos_dirty
```

```python
dirty[whos_dirty] = 0 # 모든 음수값(True)을 0으로 바꿈
dirty
```

논리 연산자 | or, & and, - not 결합하여 boolean indexing 사용


```python
linear = np.arange(-1, 1.1, 0.2)
linear
```

```python
# -0.5 ~ 0.5 사이의 값 -> 0으로 변경
whos_linear = (linear <= 0.5) & (linear >= -0.5) # 괄호를 친 것 먼저 실행
whos_linear
```

```python
linear[whos_linear] = 0
linear
```
