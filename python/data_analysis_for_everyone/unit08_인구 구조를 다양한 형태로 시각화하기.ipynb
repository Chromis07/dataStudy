지역명을 입력받아서 연령별 인구수 데이터 시각화하기

```python
import matplotlib.pyplot as plt
import csv
f = open('age.csv', encoding='cp949')
data = csv.reader(f)

result = []

# 지역명을 입력받는 변수 선언
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')


for row in data :
  if name in row[0] :
    for i in row[3:] :
      result.append(int(i.replace(',',''))) # 문자열을 정수로 변환한 뒤 result에 담기



plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')
plt.title(name + ' 지역의 인구 구조')
plt.plot(result)
plt.show()
```

항아리 모양 그래프 그리기 성별 데이터 저장하기 (Type 1)

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)
m = []
f = []

for row in data :
  if '일산서구 대화동' in row[0] :
    for i in range(0,101) :
      m.append(int(row[i+3]))
      f.append(int(row[-(i+1)])) # 100세에서부터 담는다
      
f.reverse() # 반대로 담겼던 데이터를 뒤집어 넣기

print(m)
print(f)
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)
m = []
f = []

for row in data :
  if '일산서구 대화동' in row[0] :
    m = [int(i) for i in row[3:104]] # 리스트 내포로 만들기
    for i in row[106:] : # for문으로 하기
      f.append(int(i))
print(m)
print(f)
```

```python
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')
plt.barh(range(101) ,m)
plt.barh(range(101) ,f)
plt.show()
```

항아리 그래프로 그리기 위해 남성 데이터 음수로 바꾸기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)
m = []
f = []

for row in data :
  if '일산서구 대화동' in row[0] :
    m = [-int(i) for i in row[3:104]] # 음수만 주면 간단!
    for i in row[106:] :
      f.append(int(i))

# 음수 주긴 했지만, 오류 때문에 음수가 안 보임 그래서 아래 음수 보정도 해줘야함
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

plt.figure(dpi=200)
plt.style.use('ggplot')
plt.title('일산서구 대화동의 남녀 성별 인구 분포')
plt.rc('font', family = 'NanumBarunGothic')
plt.barh(range(101) ,m, color='b', label = '남성')
plt.xlabel('인구수')
plt.ylabel('나이')
plt.barh(range(101) ,f, color='r', label = '여성')
plt.legend()
plt.show()
```
