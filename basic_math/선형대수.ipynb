## **벡터의 내적과 놈**# (norm)

내적 : 벡터끼리의 곱 <br>
$\vec{a} = (a_1,a_2,...,a_n)$ <br>
$\vec{b} = (b_1,b_2,...,b_n)$ <br> <br>
$\vec{a}*\vec{b} = (a_1,a_2,...,a_n) * (b_1,b_2,...,b_n)$ <br>
$=(a_1*b_1+a_2*b_2+...+a_n*b_n) $ <br>
$=\sum_{k=1}^n a_kb_k$

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([3, 2, 1])

print("--- dot()함수 ---")
print(np.dot(a, b))  # dot()함수에 의한 내적 
print("--- 곱의 총합 ---")
print(np.sum(a * b))  # 곱의 총합에 의한 내적
```

놈 : 벡터의 크기를 나타내는 양  [놈이란 무엇인가?](https://bskyvision.com/825)

$L^2$놈 <br>
$||\vec{x}||_2 = \sqrt{x_1^2+x_2^2+...+x_n^2} = \sqrt{\sum_{k=1}^n x_k^2} $   <br><br>
$L^1$놈 <br>
$||\vec{x}||_1 = |x_1|+|x_2|+...+|x_n| = \sum_{k=1}^n |x_k| $ <br><br>
일반화된 놈 <br>
$||\vec{x}||_p = (x_1^p+x_2^p+...+x_n^p)^\frac{1}{p} = (\sum_{k=1}^n x_k)\frac{1}{p} $


linalg.norm() 놈 구현 <br>
벡터 $\vec{a}$와 벡터 $\vec{b}$의 내적, 벡터 $\vec{a}$의 $L^2$놈과 $L^1놈$을 구하기

```python
import numpy as np

a = np.array([1, 1, -1, -1])

print("--- L2놈 ---")
print(np.linalg.norm(a))  # L2놈(디폴트)
print("--- L1놈 ---")
print(np.linalg.norm(a, 1))  # L1놈
```

```python
import numpy as np

a = np.array([1, -2, 2])
b = np.array([2, -2, 1])

print("--- 내적 ---")
print(np.dot(a, b))

print("--- L2놈 ---")
print(np.linalg.norm(a))

print("--- L1놈 ---")
print(np.linalg.norm(a, 1))
```

##**전치** (transpose)

전치 : 행과 열을 바꾸기 <br>
$A^T$

```python
import numpy as np

a = np.array([[1, 2, 3],
              [4, 5, 6]])  # 행렬

print(a.T)  # 배열명 뒤에 .T를 붙여서 전치
```

두 행렬의 열과 행이 일치하지 않으면 행렬곱이 불가능하지만<br>
전치를 하면서 가능할 수도 있음

```python
import numpy as np

a = np.array([[0, 1, 2],
             [1, 2, 3]])  # 2×3의 행렬
b = np.array([[0, 1, 2],
             [1, 2, 3]])  # 2×3의 행렬 이대로는 행렬곱을 할 수 없다

# print(np.dot(a, b))  # 전치하지 않고 행렬곱을 취하면 에러
print(np.dot(a, b.T))  # 전치에 의해 행렬곱이 가능
```

## **선형변환**

```python
import numpy as np
import matplotlib.pyplot as plt


def arrow(start, size, color):
  '''화살표를 그리는 함수'''
    plt.quiver(start[0], start[1], size[0], size[1], angles="xy",
               scale_units="xy", scale=1, color=color)

# 화살표의 시작점
s = np.array([0, 0])  # 원점

# 벡터
a = np.array([2, 3])  # 세로 벡터를 나타낸다

arrow(s, a, color="black")

# 그래프 표시
plt.xlim([-3,3])  # x의 표시 범위
plt.ylim([-3,3])  # y의 표시 범위
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.gca().set_aspect("equal")  # 가로세로비를 같게
plt.show()
```

선형변환

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([2, 3])  # 변환 전의 벡터

A = np.array([[2, -1],
              [2, -2]])

b = np.dot(A, a)  # 선형변환

print("변환 전의 벡터（a）:", a)
print("변환 후의 벡터（b）:", b)
      
def arrow(start, size, color):
    plt.quiver(start[0], start[1], size[0], size[1], angles="xy", scale_units="xy", scale=1, color=color)

s = np.array([0, 0])  # 원점

arrow(s, a, color="red")
arrow(s, b, color="blue")

# 그래프 표시
plt.xlim([-3,3])  # x의 표시 범위
plt.ylim([-3,3])  # y의 표시 범위
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.gca().set_aspect("equal")  # 가로세로비를 같게
plt.show()
```

표준기저

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([2, 3])
e_x = np.array([1, 0])  # 표준기저
e_y = np.array([0, 1])  # 표준기저

print("a:", a)
print("e_x:", e_x)
print("e_y:", e_y)
      
def arrow(start, size, color):
    plt.quiver(start[0], start[1], size[0], size[1], angles="xy",
                scale_units="xy", scale=1, color=color)

s = np.array([0, 0])  # 원점

arrow(s, a, color="blue")
arrow(s, e_x, color="g")
arrow(s, e_y, color="g")

# 그래프 표시
plt.xlim([-3,3])  # x의 표시 범위
plt.ylim([-3,3])  # y의 표시 범위
plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()
plt.gca().set_aspect("equal")  # 가로세로비를 같게
plt.show()
```

## **고윳값과 고유벡터**

linalg.eig() 함수 고유값, 고유벡터

```python
import numpy as np

a = np.array([[3, 1],
              [2, 4]])

ev = np.linalg.eig(a)  # 고유값과 고유 벡터를 동시에 구한다

print(ev[0])  # 첫 요소는 고유값 

print()

print(ev[1])  # 다음 요소는 고유 벡터
```

##**코사인 유사도**

```python
import numpy as np

def cos_sim(vec_1, vec_2):
    return np.dot(vec_1, vec_2) / (np.linalg.norm(vec_1) * np.linalg.norm(vec_2))  

a = np.array([2, 2, 2, 2])
b = np.array([1, 1, 1, 1])  # a와 같은 방향
c = np.array([-1, -1, -1, -1])  # a와 반대 방향

print("--- a와 b의 코사인 유사도 ----")
print(cos_sim(a, b))

print("--- a와 c의 코사인 유사도 ----")
print(cos_sim(a, c))
```
