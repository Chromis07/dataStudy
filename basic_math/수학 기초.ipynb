거듭제곱 그래프

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x) :
  a = 3
  return x**a

x = np.linspace(0, 2)
y = my_func(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

제곱근 그래프

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return np.sqrt(x)  # x의 양의 제곱근. x**(1/2)로도 같음

x = np.linspace(0, 9)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

$3x^2 + 2x^2 + 1$ 그래프

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):  
    return 3*x**2 + 2*x + 1

x = np.linspace(-2, 2)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

$4x^3 + 2x^2 + x + 3$ 그래프

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):  
    return 4*x**3 + 2*x**2 + x + 3

x = np.linspace(-2, 2)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

$y=x^3 - 2x^2 + 3x - 4$ 그래프

```python
import numpy as np
import matplotlib.pyplot as plt

def my_func(x):  
    return x**3 - 2*x**2 - 3*x + 4

x = np.linspace(-2, 2)
y = my_func(x)  # y = f(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```

sin 함수, cos 함수 그래프

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi)  # -π부터 π(라디안)까지
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos")
plt.legend()

plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.show()
```

0~1 사이의 소수를 균등한 확률로 반환

abs() 절댓값 

```python
import numpy as np

x = [-5, 5, -1.28, np.sqrt(5), -np.pi/2]  # 여러 값을 리스트에 저장한다
print(np.abs(x))  # 절대값을 구한다
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi)  # -π부터 π(라디안)까지
y_sin = np.abs(np.sin(x))  # sin() 함수의 절대값을 취한다
y_cos = np.abs(np.cos(x))  # cos() 함수의 절대값을 취한다

plt.scatter(x, y_sin, label="sin")
plt.scatter(x, y_cos, label="cos")
plt.legend()

plt.xlabel("x")
plt.ylabel("y")
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4)
y = np.abs(x**2 - 4)  # 이 2차함수의 절대값을 취한다 

plt.scatter(x, y)

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```
