연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기


```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

result = [] # 남성 인구 - 여성 인구 값 담을 리스트 변수 선언

name = input('성별 분포 구조가 알고 싶은 지역의 이름을 입력하세요 : ')

for row in data :
  if name in row[0] :
    for i in range(3,104) :
      result.append(int(row[i].replace(',', '')) - int(row[i+103].replace(',','')))
    break

plt.figure(dpi=100)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')
plt.title(name+' 지역의 연령별 남녀 인구 수 차이')
plt.xlabel('나이')
plt.ylabel('인구 차이')
plt.bar(range(101),result)
plt.show()
```

위치, 크기가 서로 다른 100개의 점 만들기

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for i in range(100) :
  x.append(random.randint(50,100))
  y.append(random.randint(50,100))
  size.append(random.randint(10,100))

plt.figure(dpi=200)
plt.style.use('ggplot')
plt.scatter(x, y, s= size, c=size, cmap = 'jet')
plt.colorbar()
plt.show()
```

```python
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.scatter(x, y, s= size, c=size, cmap = 'jet', alpha=0.8) # 알파 투명도
plt.colorbar()
plt.show()
```

제주도의 연령대별 성별 비율을 산점도로 표현하기

남녀 연령별 인구 데이터 scatter() 함수로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []

name = input('성별 분포 구조가 알고 싶은 지역의 이름을 입력하세요 : ')

for row in data :
  if name in row[0] :
    for i in row[3:104]:
      m.append(int(i.replace(',','')))
    for i in row[106:]:
      f.append(int(i.replace(',','')))
    break

# 지역명을 입력받는 변수 선언

plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')
plt.title(name+'의 성별 분포')
plt.scatter(m,f) # x 축은 남자, y 축은 여자
plt.show()
```

```python
plt.figure(dpi=200)
plt.scatter(m,f, c=range(101), cmap='jet', alpha=0.5)
plt.plot(range(max(m)), range(max(m)), 'g')
plt.show()
```

원 사이즈에 제곱근에 따라 넣기

```python
import csv
import matplotlib.pyplot as plt
import math

f = open('gender.csv', encoding='cp949')
data = csv.reader(f)

m = []
f = []
size = [] # 인구수 사이즈 리스트 변수 선언

name = input('성별 분포 구조가 알고 싶은 지역의 이름을 입력하세요 : ')

for row in data :
  if name in row[0] :
    for i in row[3:104]:
      m.append(int(i.replace(',','')))
      size.append(math.sqrt(int(i.replace(',','')))) # 제곱근에 따라 사이즈
    for i in row[106:]:
      f.append(int(i.replace(',','')))
    break

#인구수와 sqrt(인구수) 차이 확인
print(m)
print(size)

# 지역명을 입력받는 변수 선언

plt.figure(dpi=200)
plt.style.use('ggplot')
plt.rc('font', family = 'NanumBarunGothic')
plt.title(name+'의 성별 분포')
plt.scatter(m,f, s=size, c =range(101), cmap='jet', alpha=0.7) # x 축은 남자, y 축은 여자
plt.plot(range(max(m)), range(max(m)), 'g')
plt.colorbar()
plt.show()
```
