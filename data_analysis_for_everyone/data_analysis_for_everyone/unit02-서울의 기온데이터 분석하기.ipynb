```python
import csv

f = open("/content/seoul.csv" , 'r', encoding='cp949')

data = csv.reader(f)
header = next(data) # 첫 번째 행을 읽은 후 커서를 다음 행으로 옮기는 함수
print(header)

# 데이터를 한 줄씩 출력    
for row in data :
  print(row)

f.close()
```
