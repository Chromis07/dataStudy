CSV 파일 다루기

statistics 모듈을 사용해서 나이 변수의 평균과 표준편차 구하기

데이터 가져오기



```python
import csv

with open ("/content/Demographic_Statistics_By_Zip_Code.csv") as infile :
  data = list(csv.reader(infile))
```

```python
# 첫번째 레코드에 COUNT PARTICIPANTS 데이터 인덱스 값 추출 data[0] 컬럼 값
countParticipantsIndex = data[0].index("COUNT PARTICIPANTS")
countParticipantsIndex
```

```python
# COUNT PARTICIPANTS 데이터 추출
# 명시적으로 코딩을 해라 (데이터 믿지 마라!)
countParticipantsIndex = [ int(row[countParticipantsIndex]) for row in data[1:]]
countParticipantsIndex
```

나이의 평균과 표준편차

```python
# 파이썬 모듈보다는 판다스 모듈을 더 사용함
import statistics

print(statistics.mean(countParticipantsIndex), statistics.stdev(countParticipantsIndex))
```

nltk 에서 영단어 온톨로지(wordnet) 사용하기

```python
# https://frhyme.github.io/python-lib/nltk-wordnet/
nltk.download('wordnet')
```

```python
wn = nltk.corpus.wordnet # 코퍼스 리더(reader)

#synset : 유의어 묶음으로 단어-품사-순번 으로 구성
wn.synsets('cat')
```

```python
print(wn.synset('cat.n.01').hypernyms()) # hypernyms : 상위어
print(wn.synset('cat.n.01').hyponyms()) # hyponyms : 하위어
```

```python
print(wn.synset('guy.n.01').hypernyms())
print(wn.synset('guy.n.01').hyponyms())
```

WordNet 사용해서 synset 간 의미론적 유사도 계산하기(0-1 사이 실수)
0 이면 두 단어 서로 관계 없음
1 이면 완전한 유의어

```python
# 고양이, 링스 : 시라소니 유사도 계산
x = wn.synset('cat.n.01') # 고양이
y = wn.synset('lynx.n.01') # 시라소니

# x, y 유사도 계산
x.path_similarity(y)
```

```python
x = wn.synset('cat.n.01') # 고양이
y = wn.synset('domestic_cat.n.01') # 집고양이

# x, y 유사도 계산
x.path_similarity(y)
```

```python
x = wn.synset('cat.n.01') # 고양이
y = wn.synset('dog.n.01') # 강아지

# x, y 유사도 계산
x.path_similarity(y)
```

```python
print(wn.synset('dog.n.01').hypernyms())
print(wn.synset('dog.n.01').hyponyms())
```

자연어 처리 정규화

```python
# 1. 토큰화 (텍스트 -> 단어로 쪼갠다.)

from nltk.tokenize import WordPunctTokenizer
word_punct = WordPunctTokenizer()

text = "}Help! :))) :[ ..... :D{"

# Tokenizer는 모든 구두점(punctuation; 문장부호)을 기준으로 분리
# -> 이모티콘을 이용한 감성 분석 등 문장구조 깊이 분석 할 때 사용

word_punct.tokenize(text)
```

```python
nltk.download('punkt')
```

```python
# 단어 토크나이저
nltk.word_tokenize(text)
```

```python
# 2. 단어의 대소문자를 통일한다 (전부 다 대문자 or 소문자)
# 3. 불용어 제거 (stopwords 리스트 참조 THE 같은 것들 제거)
# 4. 형태소 분석(stemming) 단어를 형태소로 변환

# 포터 형태소 분석기 (보수적)
pstemmer = nltk.PorterStemmer()
pstemmer.stem("wonderful")
```

```python
# 5. 원형 추출
lemmatizer = nltk.WordNetLemmatizer()
lemmatizer.lemmatize("wonderful")
```

index.html 파일에서(불용어 제외) 가장 많이 등장한 단어 원형 찾아보기

```python
# www.networksciencelab.com 사이트에서 하이퍼 링크 정보 모두 가져오기

from bs4 import BeautifulSoup
from urllib.request import urlopen

# https:// << 프로토콜 무조건 넣어야 함!!
with urlopen("http://www.networksciencelab.com/") as doc :
  soup = BeautifulSoup(doc)

# 하이퍼링크 추출
links = [(link.string, link["href"]) for link in soup.find_all("a") if link.has_attr("href")]
links
```

```python
# 책 제목만 추출 (리스트 형태)
html_text = [i[0] for i in links]
html_text
```

```python
# None 데이터 제거 (join으로 하나의 텍스트 형태로 변환)
html_text = " ".join(html_text[2:])
html_text
```

```python
# 하나의 텍스트 -> index.html 파일로 저장
html_file = open("index.html", 'w')
html_file.write(html_text)
html_file.close()
```

index.html 파일에서(불용어 제외) 가장 많이 등장한 단어 원형 찾아보기

```python
from bs4 import BeautifulSoup
from collections import Counter # 빈도가 높은 숫자 세기
from nltk.corpus import stopwords # 불용어
from nltk import LancasterStemmer # 원형어

# 형태소 분류기를 생성한다.
ls = nltk.LancasterStemmer()

# 파일을 읽고 soup을 만든다.
with open("/content/index.html") as infile :
  soup = BeautifulSoup(infile)
```

```python
soup
```

```python
soup.text
```

```python
# 1) 텍스트 추출 -> 토큰화
words = nltk.word_tokenize(soup.text)
words
```

```python
# 2) 단어를 소문자로 변환
words = [w.lower() for w in words]
words
```

```python
nltk.download('stopwords')
```

```python
# 3) 불용어, 특수기호) 제거하고 단어의 형태소를 추출
# 3-1) 불용어 제거 stopwords.words("english")
# 3-2) 특수기호 제거 isalnum() - 문자열이 알파벳([a-zA-Z])과 숫자([0-9])로만 구성
# 3-3) 
words = [ls.stem(w) for w in words if w not in stopwords.words("english") and w.isalnum()]
words
```

```python
# 4) 가장 빈번하게 등장하는 단어 5개 추출
freqs = Counter(words)
print(freqs.most_common(5))
```
