rate = 유임승차 인원 / 무임승차 인원

```python
import csv
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

mx = 0 # 최대 유임승차 비율
rate = 0 # 유임승차 비율

for row in data :
  for i in range(4,8) : # 4,5,6,7 인덱스 데이터 정수형으로 변경
    row[i] = int(row[i])
  if row[4] != 0 and row[6] != 0: # 0으로 되어있는 데이터 넘기기
    rate = row[4] / row[6] # 유임승자 / 무임승차
    if rate > mx :
      mx =rate
      print(row, round(rate,2))

print(mx)
```

비율계산 방식 변경 rate = 유임승차 인원 / 전체(유임+무임) 인원

```python
import csv
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

mx = 0 # 최대 유임승차 비율
rate = 0 # 유임승차 비율

for row in data :
  for i in range(4,8) : # 4,5,6,7 인덱스 데이터 정수형으로 변경
    row[i] = int(row[i])
  if row[4] != 0 and row[6] != 0: # 0으로 되어있는 데이터 넘기기
    rate = row[4] / (row[4] + row[6]) # 유임승자 / 전체(유임+무임)
  if rate > mx :
    mx = rate
    mx_station = row[3] + ' ' + row[1] # 지하철 역 + 호선 정보 ex) 서울역 1호선
    print(mx_station, round(mx*100,2)) # 퍼센테이지 표시
```

유무임 승하자 인원이 가장 많은 역 찾기

```python
import csv
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

mx = [0]*4 # 최대 유임승차, 유임하자, 무임승차, 무임하차 인원 수
mx_station = [''] * 4 # 최대 유임승차, 유임하자, 무임승차, 무임하차 역 정보

label = ['유임승차', '유임하차', '무임승차', '무임하자']

for row in data :
  for i in range(4,8) : # 4,5,6,7 인덱스 데이터 정수형으로 변경
    row[i] = int(row[i])
    if row[i] > mx[i-4] : # 먼저 mx : 유임승차부터 비교해야하니 i-4
    # row[4] > mx[0] 유임승차 비교, row[5] > mx[1] 유임하차 비교
      mx[i-4] = row[i]
      mx_station[i-4] = row[3] + ' ' + row[1]

for i in range(4):
  print(f'{label[i]} : {mx_station[i]}, {mx[i]}명')
```

유무임 승하차 인원이 가작 적은 역 찾기

```python
import csv
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

min = [0]*4 # 최소 유임승차, 유임하자, 무임승차, 무임하차 인원 수
min_station = [''] * 4 # 최소 유임승차, 유임하자, 무임승차, 무임하차 역 정보

label = ['유임승차', '유임하차', '무임승차', '무임하자']

for row in data :
  for i in range(4,8) : # 4,5,6,7 인덱스 데이터 정수형으로 변경
    row[i] = int(row[i])
    if min_station[i-4] == '' or row[i] <= min[i-4] : # 데이터 첫줄이 무조건 넣어줌
      min[i-4] = row[i]
      min_station[i-4] = row[3] + '역 ' + row[1]

for i in range(4):
  print(f'{label[i]} : {min_station[i]}, {min[i]}명')
```

모든 역의 유무임 승하차 비율을 파이 차트로 나타내기

```python
import csv
import matplotlib.pyplot as plt

f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

mx = [0]*4 # 최대 유임승차, 유임하자, 무임승차, 무임하차 인원 수
mx_station = [''] * 4 # 최대 유임승차, 유임하자, 무임승차, 무임하차 역 정보

label = ['유임승차', '유임하차', '무임승차', '무임하자']

for row in data :
  for i in range(4,8) : # 4,5,6,7 인덱스 데이터 정수형으로 변경
    row[i] = int(row[i])
    plt.figure(dpi=200)
    plt.title(row[3] + ' ' + row[1])
    plt.pie(row[4:8], labels=label, autopct='%.1f%%')
    plt.axis('equal')
    plt.savefig(row[3]+ ' ' +row[1]+'.png') # 이미지 저장
    plt.show()
```
