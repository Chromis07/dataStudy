step 3) 최고 기온과 최고 기온이었던 날짜 찾기

```python
import csv

max_temp = -999 # 최고 기온 값을 저장할 변수 -> 명확한 결과를 위해 극적인 값으로 셋팅
max_date = '' # 최고 기온이 가장 높았던 날짜를 저장하는 변수

f = open("/content/seoul.csv" , 'r', encoding='cp949')

data = csv.reader(f)
header = next(data) # 헤더 추출

for row in data :
  # 최고 기온 데이터 ''(결측치) 확인
  if row[-1] == '' :
    row[-1] = -999 # -999를 넣어서 빈 문자열('')이 있던 자리라고 표시

  # 최고 기온 데이터 
  row[-1] = float(row[-1]) # 기온 데이터 실수형으로 변환

  # 최고 기온 정보를 저장

  if max_temp < row[-1] :
    max_date = row[0]
    max_temp = row[-1]

f.close()

print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은 ', max_date+'로 ', max_temp,'도 였습니다.' )
```

step 4) 최저 기온과 최저 기온이었던 날짜 찾기

```python
import csv

min_temp = 999 # 최저 기온 값을 저장할 변수 -> 명확한 결과를 위해 극적인 값으로 셋팅
min_date = '' # 최저 기온이 가장 낮았던 날짜를 저장하는 변수

f = open("/content/seoul.csv" , 'r', encoding='cp949')

data = csv.reader(f)
header = next(data) # 헤더 추출

for row in data :
  # 최저 기온 데이터 ''(결측치) 확인
  if row[3] == '' :
    row[3] = 999 # 999를 넣어서 빈 문자열('')이 있던 자리라고 표시

  # 최저 기온 데이터 
  row[3] = float(row[3]) # 기온 데이터 실수형으로 변환

  # 최저 기온 정보를 저장

  if min_temp > row[3] :
    min_date = row[0]
    min_temp = row[3]

f.close()

print('기상 관측 이래 서울의 최저 기온이 가장 낮았던 날은 ', min_date+'로 ', min_temp,'도 였습니다.' )
```

plt.plot(데이터셋) 그래프 그리기

```python
import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40]) # y 값이 주어지면, 자동으로 x 값을 0부터 1씩 증가함
plt.show
```

plt.plot(x,y) 그래프 그리기

```python
plt.title('plotting') # 타이틀 속성 추가
plt.plot([1, 2, 3, 4],[12, 43, 33, 51]) # plt.plot(x ,y) 함수에 데이터 입력하기
plt.show
```

```python
plt.title('plotting') # 타이틀 속성 추가
plt.plot([10, 20, 30, 40], color='skyblue', label = 'asc') # 증가를 의미하는 asc 범례
plt.plot([40, 30, 20, 10], 'pink', label = 'desc') # 감소를 의미하는 desc 범례

plt.legend(loc=6) # 그래프에 범례 속성 추가
plt.show
```

```python
plt.title('plotting') # 타이틀 속성 추가
plt.plot([10, 20, 30, 40], color='r', linestyle = '--', label = 'dash') # 기본 색상은 약어로 표시 가능
plt.plot([40, 30, 20, 10], 'g', label = 'green')

plt.legend(loc=6) # 그래프에 범례 속성 추가
plt.show
```
