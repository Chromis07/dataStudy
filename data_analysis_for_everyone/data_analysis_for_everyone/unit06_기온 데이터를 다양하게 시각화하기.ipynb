```python
import random

def diceroll(n) :
  '''주사위를 n번 굴려서 리스트로 저장해두는 함수'''
  dice_list = []
  for i in range(n) :
    dice_list.append(random.randint(1,6))
  return dice_list

# 주사위 굴린 걸 히스토그램으로 그리기
import matplotlib.pyplot as plt

plt.figure(dpi=200) # 해상도
plt.hist(diceroll(1000000), bins=6) 
plt.show()
```

최고 기온 데이터를 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 다음 줄부터

result = [] # 최고 기온 데이터를 저장할 리스트 생성

for row in data :
  if row[-1] != '' : # 최고 기온 데이터 값이 존재하면 (결측치 제외)
    result.append(float(row[-1]))

# 최고 기온 히스토그램
plt.figure(dpi=200)
plt.hist(result, bins=100, color = 'r')
plt.show
```

1월과 8월의 데이터를 히스토그램으로 시각화하기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 다음 줄부터

aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언
jan = [] # 1월의 최고 기온을 담을 jan 리스트 변수 선언

for row in data :
  month = row[0].split('-')[1] # 날짜 데이터에서 월 데이터 추출
  if row[-1] != '' :
    if month == '08' :
      aug.append(float(row[-1]))
    if month == '01' :
      jan.append(float(row[-1]))

plt.figure(dpi=200)
plt.hist(aug, bins=50, color = 'g', label='aug')
plt.hist(jan, bins=50, color = 'skyblue', label='jan')
plt.legend()
plt.show
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data) # 헤더 다음 줄부터

aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언
jan = [] # 1월의 최고 기온을 담을 jan 리스트 변수 선언
apr = [] # 4월의 최고 기온을 담을 june 리스트 변수 선언

for row in data :
  month = row[0].split('-')[1] # 날짜 데이터에서 월 데이터 추출
  if row[-1] != '' :
    if month == '08' :
      aug.append(float(row[-1]))
    if month == '01' :
      jan.append(float(row[-1]))
    if month == '04' :
      apr.append(float(row[-1]))

plt.figure(dpi=200)
plt.hist(aug, bins=50, color = 'g', label='aug')
plt.hist(jan, bins=50, color = 'skyblue', label='jan')
plt.hist(apr, bins=50, color = 'purple', label='apr')
plt.xlabel('temperture')
plt.ylabel('count')
plt.legend()
plt.show
```
